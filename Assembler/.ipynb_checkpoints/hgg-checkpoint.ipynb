```python
#imports
##########################################################################
# for the argumnets
import sys
# ordered dictionary
from collections import OrderedDict
# for colored printing
from colorama import Fore, Back, Style
##########################################################################
```

```python
# getting the file name
fileName = sys.argv[1]
# setting the default outputname
fileNameOut = 'a.mem'
# check whether to change the name or not
if (argumentLenght == 3):
    fileNameOut = sys.argv[2]
```

```python
#compile the file
def getInput(file, arrayISA):
    # loop on line and neglect empty lines
    for line in file:
        if(line.strip()):
            if(';' in line):
                # to delte any commnets form the line
                indexofhash = line.rfind(';')
                line = line[0:indexofhash]
                if(not line.strip()):
                    continue
            #add line to the array
            arrayISA.append(line)
    print(arrayISA)
```

```python
def CompileOutput(arrayISA, out):
    reg = {
        'R0' : 0b000,
        'R1' : 0b001,
        'R2' : 0b010,
        'R3' : 0b011,
        'R4' : 0b100,
        'R5' : 0b101,
        'R6' : 0b110,
        'R7' : 0b111,
    }
    opCode = {
        'NOP' : 0b0000
        'group0' : 0b0001
        'group1' : 0b0010
        'group2' : 0b0011
        'b_type' : 0b0100
        'c_type' : 0b0101
        's_type' : 0b0110
        'm_type' : 0b0111
        'I_type' : 0b1000
        'CLR_FLAGS' : 0b1001
        'CALL' : 0b1010
    }
    func = {
        'funct0' : 0b00
        'funct1' : 0b01
        'funct2' : 0b10
        'funct3' : 0b11
    }
    rdsShift = 3
    rds2Shift = 1
    rsrShift = 2
    opCodeShift = 3
    numISA = 0
    currentIP = 32;
    for line in arrayISA:
        if (currentIP > 1048575 or currentIP < 0):
            return 'out of bound in writing memory'
        print(line)
        lineOld = line
        line = line.replace(',', ' ')
        line = line.split()
        temp = 0b0000000000000000
        print(len(line))
        if(line[0].casefold() == 'DIV'.casefold()):
            if(len(line) == 3):
                op1 = reg.get(line[1], None)
                op2 = reg.get(line[2], None)
                if(op1 == None or op2 == None):
                    return 'Error in ' + lineOld + 'Instruction number:' + numISA
                temp = temp | (opCode['group0'] << opCodeShift) | (op2 << rdsShift) | (op1 << rsrShift) | (func['funct0'])
                out[currentIP] = temp
            else:
                return 'Error in ' + lineOld + 'Instruction number:' + numISA
            
        else if(line[0].casefold() == 'MUL'.casefold()):#note there is 2 r des types
            if(len(line) == 4):
                op1 = reg.get(line[1], None)
                op2 = reg.get(line[2], None)
                op3 = reg.get(line[3], None)
                if(op1 == None or op2 == None or op3 == None):
                    return 'Error in ' + lineOld + 'Instruction number:' + numISA
                temp = temp | (opCode['group0'] << opCodeShift) | (op2 << rdsShift) | (op1 << rsrShift) | (op3 << rds2Shift) | (func['funct1'])
                out[currentIP] = temp
            else:
                return 'Error in ' + lineOld + 'Instruction number:' + numISA
            
        else if(line[0].casefold() == 'ADD'.casefold()):
        else if(line[0].casefold() == 'SUB'.casefold()):
        else if(line[0].casefold() == 'SHL'.casefold()):
        else if(line[0].casefold() == 'SHR'.casefold()):
        else if(line[0].casefold() == 'INC'.casefold()):
        else if(line[0].casefold() == 'DEC'.casefold()):
        else if(line[0].casefold() == 'OR'.casefold()):
        else if(line[0].casefold() == 'AND'.casefold()):
        else if(line[0].casefold() == 'NOT'.casefold()):
        else if(line[0].casefold() == 'MOV'.casefold()):
        else if(line[0].casefold() == 'JZ'.casefold()): #note there is two types
        else if(line[0].casefold() == 'JN'.casefold()):
        else if(line[0].casefold() == 'JC'.casefold()):
        else if(line[0].casefold() == 'JMP'.casefold()): # rds and imediate
        else if(line[0].casefold() == 'CLRZ'):
        else if(line[0].casefold() == 'CLRN'.casefold()):
        else if(line[0].casefold() == 'CLRC'.casefold()):
        else if(line[0].casefold() == 'CLRINT'.casefold()):
        else if(line[0].casefold() == 'SETZ'.casefold()):
        else if(line[0].casefold() == 'SETN'.casefold()):
        else if(line[0].casefold() == 'SETC'.casefold()):
        else if(line[0].casefold() == 'SETINT'.casefold()):
        else if(line[0].casefold() == 'PUSH'.casefold()): # rds pc_flags pc
        else if(line[0].casefold() == 'POP'.casefold()): # rds pc_flags pc
        else if(line[0].casefold() == 'OUT'.casefold()):  #rds, portnumber
        else if(line[0].casefold() == 'IN'.casefold()):    #rds, portnumber
        else if(line[0].casefold() == 'LDD'.casefold()):
        else if(line[0].casefold() == 'STD'.casefold()):
        else if(line[0].casefold() == 'LDM'.casefold()):
        else if(line[0].casefold() == 'CALL'.casefold()):
#         else if(line[0].casefold() == 'JMP'.casefold()):
#         else if(line[0].casefold() == 'JC'.casefold()):
#         else if(line[0].casefold() == 'JN'.casefold()):
#         else if(line[0].casefold() == 'JZ'.casefold()):
        else if(line[0].casefold() == 'CLR_FLAGS'.casefold()):
        else if(line[0].casefold() == 'NOP'.casefold()):
            if(len(line) == 1):
                out[currentIP] = temp
            else:
                return 'Error in ' + lineOld + 'Instruction number:' + numISA 
        else if(line[0].casefold() == 'CALL'.casefold()): # rds , imediate
        else if(line[0].casefold() == 'RET'.casefold()):
        else if(line[0].casefold() == 'RTI'.casefold()):
        else:
            return 'Invaild Instruction (instruction is not in the ISA please check your code)'
        currentIP += 1
        numISA += 1
    return 0
```

```python
arrayISA = []
# try opening the file
try:
    file = open('test.txt', 'r')
#     compileOutput(file, fileNameOut)
    getInput(file, arrayISA)
    if(len(arrayISA) == 0):
        print(Fore.RED + 'File is Empty! Please check it again.')
        print(Style.RESET_ALL)
        exit()
    # Closing the file
    file.close()

#   testing
    print(arrayISA)

except IOError:
    print(Fore.RED + 'Could not open file! Please check it again.')
    print(Style.RESET_ALL)
    exit()
    
    
 
```
